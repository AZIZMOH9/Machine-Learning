```python
import pandas as pd
import numpy as np
```

```python
movie=pd.read_csv('movies.csv')
movie.head()
```

```python
raiting=pd.read_csv('ratings.csv')
raiting.head()
```

```python
df=pd.merge(raiting,movie,on="movieId").drop(["genres","timestamp"],axis=1)
```

```python
df.head()
```

```python
df=df.pivot_table(index='userId',columns='title',values='rating')
```

```python
df.head(30)
```

```python
df=df.dropna(thresh=10,axis=1)
```

```python
df=df.fillna(0)
```

```python
def get_similar(mov,rai):
    score=df[mov]*(rai-2.5)
    score=score
    score=score.sort_values(ascending=False)[1:]
    return    score
```

```python
df=df.corr('pearson')
```

```python
df.head()
```

```python
get_similar("Star Wars: Episode I - The Phantom Menace (1999)",5)
```

```python
set=[("Star Wars: Episode I - The Phantom Menace (1999)",5),("Star Wars: Episode II - Attack of the Clones (2002)",4)]
```

```python
similar_=pd.DataFrame()
for i,v in set:
    similar_=similar_.append(get_similar(i,v),ignore_index=True)
```
