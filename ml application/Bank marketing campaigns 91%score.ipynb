# Feature Engineering

### according to data visualization that i did above im going to lable and encoded some features

```python
data=pd.get_dummies(data=data,columns=['default'],drop_first=True)
```

```python
poutcome={'failure':0,'nonexistent':1,'success':2}
marital={'single':3,'married':2,'divorced':1,'unknown':0}
loan={'unknown':0,'no':2,'yes':1}
contact={'telephone':1,'cellular':0}
housing={'yes':2,'no':1,'unknown':0}
education={'university.degree':7,'high.school':6,'professional.course':5,'basic.9y':5,'unknown':4,'basic.6y':3,'basic.4y':4,'illiterate':3}
job={'unknown':0,'housemaid':0,'entrepreneur':1,'self-employed':1,'retired':2,'student':2,'services':5,'blue-collar':3,'technician':3,'admin.':4,'management':2, 'unemployed':3}
empvarrate={'-3.4':2,'-3':2,'-2.9':2,'-1.7':3,'-1.1':3,'-1.8':2,'1.4':2,'1.4':1,'-1.8':1}
```

```python
data['poutcome']=data['poutcome'].map(poutcome)
data['marital']=data['marital'].map(marital)
data['loan']=data['loan'].map(loan)
data['contact']=data['contact'].map(contact)
data['housing']=data['housing'].map(housing)
data['education']=data['education'].map(education)
data['job']=data['job'].map(job)
```

```python
data.info()
```

### spreating data

```python
x=data.drop('y_yes',axis=1)
y=data['y_yes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train=pd.get_dummies(data=X_train,columns=['day_of_week','month'],drop_first=True)
X_test=pd.get_dummies(data=X_test,columns=['day_of_week','month'],drop_first=True)
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
X_train[['age','duration','pdays','cons.price.idx','cons.conf.idx','nr.employed']]=s.fit_transform(X_train[['age','duration','pdays','cons.price.idx','cons.conf.idx','nr.employed']])
```

```python
X_test[['age','duration','pdays','cons.price.idx','cons.conf.idx','nr.employed']]=s.transform(X_test[['age','duration','pdays','cons.price.idx','cons.conf.idx','nr.employed']])
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
l=LogisticRegression(C=0.5)
```

```python
l.fit(X_train,y_train)
```

```python
p=l.predict(X_test)
```

```python
l.score(X_test,y_test) #accuracy
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
classification_report(y_test,p)
```

```python
confusion_matrix(y_test,p)
```
